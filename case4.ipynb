```python
def correctCountRub(rub):
    if rub < 0:
        return "Введите корректное значение"
    words = ["рубль", "рубля", "рублей"]
    lastNum = rub % 100
    if 10 < lastNum < 20:
        return str(rub) + " " + str(words[2]) 
    else:
        lastNum %= 10
        if 0 < lastNum < 2:
            return str(rub) + " " + str(words[0]) 
        elif 1 < lastNum < 5:
            return str(rub) + " " + str(words[1]) 
        else:
            return  str(rub) + " " + str(words[2]) 
```

```python
for ruble in range(-1, 10000):
    print(correctCountRub(ruble))
```

```python
inputRub = int(input())
correctCountRub(inputRub)
```
